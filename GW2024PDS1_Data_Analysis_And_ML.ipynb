```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_squared_error, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load Data - This is a sample dataset that simulates medical consultations
df = pd.read_csv('consultations.csv')
df.head()
```

```python
# Preprocessing
# Split BP into systolic and diastolic components
df[['systolic', 'diastolic']] = df['bp'].str.split('/', expand=True)
df['systolic'] = df['systolic'].astype(int)  # Convert systolic values to integers
df['diastolic'] = df['diastolic'].astype(int)  # Convert diastolic values to integers

# Drop non-numeric columns that won't be used in modeling
df.drop(columns=['bp', 'remarks', 'followup', 'doctor_email', 'patient_name', 'created_on'], inplace=True)

# Label Encoding for Categorical Variables - Converts categories to numeric values
le_complaints = LabelEncoder()
le_medicines = LabelEncoder()
le_doctor_name = LabelEncoder()

df['complaints'] = le_complaints.fit_transform(df['complaints'])  # Encode complaints
df['medicines'] = le_medicines.fit_transform(df['medicines'])  # Encode medicines
df['doctor_name'] = le_doctor_name.fit_transform(df['doctor_name'])  # Encode doctor names
```

```python
# Feature Selection - Select features for modeling
X = df[['complaints', 'systolic', 'diastolic', 'temperature', 'medicines']]
y_regression = df['sugar']  # Target for regression: sugar levels
y_classification = (df['systolic'] > 140) | (df['diastolic'] > 90)  # Target for classification: high blood pressure
```

```python
# Train/Test Split - Split data into training and testing sets
X_train, X_test, y_train_reg, y_test_reg = train_test_split(X, y_regression, test_size=0.2, random_state=42)
_, _, y_train_cls, y_test_cls = train_test_split(X, y_classification, test_size=0.2, random_state=42)
```

```python
# Standardize Features - Standardize features by removing the mean and scaling to unit variance
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)  # Fit and transform the training data
X_test = scaler.transform(X_test)  # Transform the testing data
```

```python
# Regression Model - Linear Regression for predicting sugar levels
regressor = LinearRegression()
regressor.fit(X_train, y_train_reg)  # Train the regression model
y_pred_reg = regressor.predict(X_test)  # Predict using the trained model

# Regression Evaluation - Evaluate the regression model
mse = mean_squared_error(y_test_reg, y_pred_reg)  # Calculate Mean Squared Error
print(f"Regression Mean Squared Error: {mse}")
```

```python
# Classification Model - Logistic Regression for predicting high blood pressure condition
classifier = LogisticRegression()
classifier.fit(X_train, y_train_cls)  # Train the classification model
y_pred_cls = classifier.predict(X_test)  # Predict using the trained model

# Classification Evaluation - Evaluate the classification model
accuracy = accuracy_score(y_test_cls, y_pred_cls)  # Calculate Accuracy
print(f"Classification Accuracy: {accuracy}")
```

```python
# Visualization
# Correlation Heatmap - Visualize correlations between features
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')  # Plot heatmap
plt.title('Feature Correlation')
plt.show()
```

```python
# Regression Results - Plot actual vs predicted sugar levels
plt.figure(figsize=(10, 6))
plt.scatter(y_test_reg, y_pred_reg)
plt.plot([y_test_reg.min(), y_test_reg.max()], [y_test_reg.min(), y_test_reg.max()], 'k--', lw=2)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Regression: Actual vs Predicted Sugar Levels')
plt.show()
```

```python
# Classification Results - Plot distribution of high blood pressure cases
sns.countplot(x=y_classification, palette='Set2')
plt.title('Classification: High Blood Pressure Distribution')
plt.show()
```

```python
# Prediction Cases - Test the models with new data

# New test cases for predictions
new_cases = pd.DataFrame({
    'complaints': ['Headache', 'Cough', 'Back Pain'],
    'systolic': [135, 115, 145],
    'diastolic': [85, 75, 95],
    'temperature': [98.7, 100.2, 97.9],
    'medicines': ['Tylenol', 'Robitussin', 'Advil']
})

# Encode new categorical data to match training data
for col, encoder in zip(['complaints', 'medicines'], [le_complaints, le_medicines]):
    new_cases[col] = new_cases[col].map(lambda x: encoder.transform([x])[0] if x in encoder.classes_ else -1)

# Handle unknown categories by mapping them to a default or removing them
new_cases = new_cases[new_cases['complaints'] != -1]  # Remove unknown complaint categories
new_cases = new_cases[new_cases['medicines'] != -1]  # Remove unknown medicine categories

# Standardize the new test cases
new_cases_scaled = scaler.transform(new_cases)

# Predict sugar levels (Regression)
predicted_sugar_levels = regressor.predict(new_cases_scaled)
print(f"Predicted Sugar Levels: {predicted_sugar_levels}")

# Predict high blood pressure condition (Classification)
predicted_bp_condition = classifier.predict(new_cases_scaled)
print(f"Predicted High Blood Pressure Condition: {predicted_bp_condition}")
```
