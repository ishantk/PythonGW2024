```python
# Which doctor prescribes the most medicines ?
top_doctor = data['doctor_name'].value_counts().idxmax()
print("Top Doctor:", top_doctor)
```

```python
# Number of Precriptions Per Day
# Converted Data Type from String to datetime
data['created_on'] = pd.to_datetime(data['created_on'])
```

```python
data['created_on'].value_counts()
```

```python
daily_prescriptions = data.groupby(data['created_on'].dt.date).size()
print(daily_prescriptions)
```

```python
daily_prescriptions.plot(kind='bar', title='Daily Prescriptions')
plt.xlabel('Date')
plt.ylabel('Number of Prescriptions')
plt.show()
```

```python
medicine_frequency = data.groupby(data['medicines']).size()
print(medicine_frequency)
```

```python
medicine_frequency.plot(kind='bar', title='Medicine Frequency')
plt.xlabel('Medicine')
plt.ylabel('Number of Medicines')
plt.show()
```

```python
monthly_prescriptions = data.groupby(data['created_on'].dt.to_period('M')).size()
yearly_prescriptions = data.groupby(data['created_on'].dt.to_period('Y')).size()

print(monthly_prescriptions)
print(yearly_prescriptions)

# Plotting monthly prescriptions
monthly_prescriptions.plot(kind='line', title='Monthly Prescriptions Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Prescriptions')
plt.show()
```
